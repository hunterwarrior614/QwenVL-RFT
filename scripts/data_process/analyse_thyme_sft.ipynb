```python
import pandas as pd

df = pd.read_parquet("../../../Thyme-SFT/data/wo_thinking_thyme_single_round-00000-of-00146.parquet",engine='fastparquet')
sample = df.iloc[10]   
# print(sample)
raw_base64 = sample['image'] # 可能是字符串或列表

if isinstance(raw_base64, list):
    # 通常第一张是原始输入图
    img_b64 = raw_base64[0]
else:
    img_b64 = raw_base64

# print(sample['question'])
print(sample['response'])
```

```python
import base64
from PIL import Image
import io


# 解码为二进制字节流
image_bytes = base64.b64decode(img_b64)

# 用 PIL 打开
image = Image.open(io.BytesIO(image_bytes))
image.show()
```
